*   **Pregunta 1.** 

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario. 

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
docs_nuevo = []
def find_n(docs):
  x = ''
  for i in docs:
    if '\n' in i:
      x = i 
      x = x.rstrip('\n')
      docs_nuevo.append(x)
  print(docs_nuevo[:10])


find_n(docs)
    
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!". 

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen. 

Indica cuántos resultados obtuviste.



```python
cadenas = ''.join(docs_nuevo)
cadenas = re.split(r"[\s]",cadenas)
#cadenas: texto como palabras
#docs_nuevo: texto como frases 
```

```python
c=0

for strs in cadenas:
  if re.findall(r"!!$", strs):
    c = c+1
    print(strs)
```

```python
print("Cantidad de palabras que terminan con dos o más signos de admiración seguidos: ", c)
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string. 

Indica la cantidad de resultados obtenidos.

```python
cantidad=[]
for comentario in docs_nuevo:
  palabras = comentario.split()[1:]
  for pal in palabras:
    if re.match(r"^[A-Z]\w+[a-z]+$",pal):
      cantidad.append(pal)
print(cantidad)
print("Cantidad de palabras que inician con mayúscula y terminan \ncon una minúscula y que no son la primera palabra del comentario:",len(cantidad))
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
caracteres = r'[.,;!?]'
minusculas = r'[A-Z]'
espacios_dobles = r'[\s]{2,}'
#iterar cada comentario en la lista 
#lambda es la funcion para convertir mayusculas a minusculas con su respectiva letra 
for comentario in range(len(docs_nuevo)):
  docs_nuevo[comentario] = re.sub(caracteres, '', docs_nuevo[comentario])
  docs_nuevo[comentario] = re.sub(minusculas,lambda x: x.group().lower(), docs_nuevo[comentario])
  docs_nuevo[comentario] = re.sub(espacios_dobles,'', docs_nuevo[comentario])
print(docs_nuevo[:10])
```

```python
corpus_frases = docs_nuevo
corpus_frases[:5]
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus. 

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus. 

```python
palabras_corpus = r"[\w']+"
t=0
for comentario in range(len(docs_nuevo)):
  docs_nuevo[comentario] = re.findall(palabras_corpus, docs_nuevo[comentario])
  t += len(docs_nuevo[comentario])
print(docs_nuevo[:10])
```

```python
print("Total de tokens obtenidos en todo el corpus:", t)
```

```python
corpus_tokenizado = docs_nuevo
corpus_tokenizado[:3]
```

Con base a la lista de stopwords que se te proporciona, realiza un proceso de limpieza eliminando todas estas palabras del corpus obtenido en el ejercicio anterior.

```python
tokens = 0
tokens_unicos = 0

for i in range(len(corpus_tokenizado)):
  filtro = [w for w in corpus_tokenizado[i] if w.lower() not in mis_stopwords]
  docs_sinstopwords = ' '.join(filtro)
  corpus_tokenizado[i] = docs_sinstopwords
  tokens += len(filtro)
  if tokens not in mis_stopwords:
    tokens_unicos +=1
print(corpus_tokenizado[:20])


```

```python
print("Total de tokens obtenidos",tokens)
print("Total de tokens únicos obtenidos", tokens_unicos)
```
